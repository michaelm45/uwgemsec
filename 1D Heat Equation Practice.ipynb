```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#all initial conditions
L = 20 #length of rod
n = 20 #number of elements
T0 = 0 #initial temperature
T1b = 40 #Temperature at boundary 1 is 40 degrees
T2b = 10 #Temperature at boundary 2 is 10 degrees
dx = L/n #change in x is the length of the rod divided by n elements
alpha = 0.5 #thermal diffusivity constant66790
t_final = 60 #number of seconds to run
dt = 0.5 #timestep
```

```python
x = np.linspace(dx/2, L-dx/2, n) #nodes at center of each element
```

```python
#vector of ones n units long 
#vector that is one row and n columns
#and each element will have length of one 
#multiplied by starting temperature

T = np.ones(n) * T0
```

```python
#each node is assigned a temp
#calculate a derivatice for each node at each
#timestep, defining vector now instead of later
#since I'm doing a for-loop later
dTdt = np.empty(n)
```

```python
#create time vector from zero to t final in
#increments of dt
t = np.arange(0, t_final, dt)
```

```python
for j in range(1,len(t)):
    for i in range(1, n-1):
          dTdt[i] = alpha*(-(T[i]-T[i-1]/dx**2)+(T[i+1]-T[i])/dx**2)
    dTdt[0] = alpha*(-(T[0]-T1b/dx**2)+(T[1]-T[0])/dx**2)
    dTdt[n-1] = alpha*(-(T[n-1]-T[n-2]/dx**2)+(T2b-T[n-1])/dx**2)
    T = T + dTdt*dt
    plt.figure(1)
    plt.plot(x,T)
    plt.axis([0, L, 0, 50])
    plt.xlabel('Distance (m)')
    plt.ylabel('Temperature (C)')
    plt.show()
```
